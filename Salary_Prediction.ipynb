```python
def scatter_y(true_y, predicted_y):
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.plot(true_y, predicted_y, '.k')
    ax.plot([0, 30], [0, 30], '--k')
    ax.plot([0, 30], [2, 32], ':k')
    ax.plot([2, 32], [0, 30], ':k')
    rms = (true_y - predicted_y).std()
    ax.text(25, 3,
            "Root Mean Square Error = %.2g" % rms,
            ha='right', va='bottom')
    ax.set_xlim(0, 30)
    ax.set_ylim(0, 30)
    ax.set_xlabel('Salary')
    ax.set_ylabel('CollegeGpa')
    return rms
```
